## 根據範例，取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta


url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='/Users/cafemato/chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

import time
for i in range(50):
    time.sleep(2)
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
today = datetime.now()

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)

for d in soup.find(class_="part_list_2").find_all('h3'):
    dt = d.find("span", class_="date").text
    title = d.find_all('a')[-1].text
    news_dt = datetime.strptime(dt, "%Y/%m/%d %H:%M")
    delta = today - news_dt
    if delta.days < 1:
        print(news_dt,title)
```

## 取出今天下午三點到五點的新聞

```python
today_pm3 = datetime.strptime("2019/12/23 15:00", "%Y/%m/%d %H:%M")
today_pm5 = datetime.strptime("2019/12/23 17:00", "%Y/%m/%d %H:%M")

for d in soup.find(class_="part_list_2").find_all('h3'):
    dt = d.find("span", class_="date").text
    news_dt = datetime.strptime(dt, "%Y/%m/%d %H:%M")
    if news_dt > today_pm3 and news_dt < today_pm5:
        print(news_dt)
```
