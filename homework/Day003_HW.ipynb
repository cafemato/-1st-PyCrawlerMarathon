### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
import os, sys
import xmltodict
fh = open("./data/64_72hr_CH.xml", "r")
xml = fh.read()
fh.close()
```

```python
# 1. 請問高雄市有多少地區有溫度資料？

d = dict(xmltodict.parse(xml))
area = d['cwbopendata']['dataset']['locations']
locations = area["location"]
print(len(locations))
'''
'''
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for data in locations:  
    print(data["weatherElement"][0]["time"][0]["dataTime"])
    print(data["weatherElement"][0]["time"][0]["elementValue"]["measures"]+data["weatherElement"][0]["time"][0]["elementValue"]["value"]+"度")
'''

'''
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
first = locations[0]
print(first["weatherElement"][0]["time"][0]["dataTime"])
for data in first["weatherElement"][0]["time"]:
    print(data["dataTime"])
    print(data["elementValue"]["measures"]+data["elementValue"]["value"]+"度")
'''

'''
```
